```python
import pandas as pd
```

```python
data=pd.read_csv("A:\\internship\\task3\\tips.csv")
```

```python
display(data.head(10))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(data["day"],data['tip'])
plt.title("Scatter Plot")
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
plt.scatter(data["day"],data['tip'],s=data['total_bill'],c=data['total_bill'],cmap='viridis')
plt.title("Scatter Plot")
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
plt.plot(data['tip'])
plt.plot(data['size'])
plt.bar(data['day'],data['tip'])
plt.title("Scatter Plot")
plt.xlabel('Day')
plt.ylabel('Tips')
plt.show()
```

```python
plt.bar(data['day'],data['tip'])
plt.title("Bar Chat")
plt.xlabel('Day')
plt.ylabel('Tips')
plt.show()
```

```python
plt.hist(data['total_bill'],bins=20,edgecolor='black')
plt.xlabel('total bill')
plt.ylabel('frequency')
plt.show()
```
